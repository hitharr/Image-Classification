```
from pyspark.ml.image import ImageSchema
from pyspark.sql.functions import lit

bug_df= ImageSchema.readImages(img_dir + "/Bug").withColumn("label", lit(0))
dark_df= ImageSchema.readImages(img_dir + "/Dark").withColumn("label", lit(1))
dragon_df= ImageSchema.readImages(img_dir + "/Dragon").withColumn("label", lit(2))
electric_df= ImageSchema.readImages(img_dir + "/Electric").withColumn("label", lit(3))
fairy_df= ImageSchema.readImages(img_dir + "/Fairy").withColumn("label", lit(4))
fighting_df= ImageSchema.readImages(img_dir + "/Fighting").withColumn("label", lit(5))
fire_df= ImageSchema.readImages(img_dir + "/Fire").withColumn("label", lit(6))
flying_df= ImageSchema.readImages(img_dir + "/Flying").withColumn("label", lit(7))
ghost_df= ImageSchema.readImages(img_dir + "/Ghost").withColumn("label", lit(8))
grass_df= ImageSchema.readImages(img_dir + "/Grass").withColumn("label", lit(9))
ground_df= ImageSchema.readImages(img_dir + "/Ground").withColumn("label", lit(10))
ice_df= ImageSchema.readImages(img_dir + "/Ice").withColumn("label", lit(11))
normal_df= ImageSchema.readImages(img_dir + "/Normal").withColumn("label", lit(12))
poison_df= ImageSchema.readImages(img_dir + "/Poison").withColumn("label", lit(13))
psychic_df= ImageSchema.readImages(img_dir + "/Psychic").withColumn("label", lit(14))
rock_df= ImageSchema.readImages(img_dir + "/Rock").withColumn("label", lit(15))
steel_df= ImageSchema.readImages(img_dir + "/Steel").withColumn("label", lit(16))
water_df= ImageSchema.readImages(img_dir + "/Water").withColumn("label", lit(17))
```

```
#split into training and test sets
bug_train, bug_test = bug_df.randomSplit([0.8,0.2])
dark_train, dark_test= dark_df.randomSplit([0.8,0.2])
dragon_train, dragon_test= dragon_df.randomSplit([0.8,0.2])
electric_train, electric_test= electric_df.randomSplit([0.8,0.2])
fairy_train, fairy_test= fairy_df.randomSplit([0.8,0.2])
fighting_train, fighting_test= fighting_df.randomSplit([0.8,0.2])
fire_train, fire_test= fire_df.randomSplit([0.8,0.2])
flying_train, flying_test= flying_df.randomSplit([0.8,0.2])
ghost_train, ghost_test= ghost_df.randomSplit([0.8,0.2])
grass_train, grass_test= grass_df.randomSplit([0.8,0.2])
ground_train, ground_test= ground_df.randomSplit([0.8,0.2])
ice_train, ice_test= ice_df.randomSplit([0.8,0.2])
normal_train, normal_test= normal_df.randomSplit([0.8,0.2])
poison_train, poison_test= poison_df.randomSplit([0.8,0.2])
psychic_train, psychic_test= psychic_df.randomSplit([0.8,0.2])
rock_train, rock_test= rock_df.randomSplit([0.8,0.2])
steel_train, steel_test= steel_df.randomSplit([0.8,0.2])
water_train, water_test= water_df.randomSplit([0.8,0.2])
```

```
#create test set
train_df = bug_train.unionAll(dark_train).unionAll(dragon_train).unionAll(electric_train).unionAll(fairy_train).unionAll(fighting_train).unionAll(fire_train).unionAll(flying_train).unionAll(ghost_train).unionAll(grass_train).unionAll(ground_train).unionAll(ice_train).unionAll(normal_train).unionAll(poison_train).unionAll(psychic_train).unionAll(rock_train).unionAll(steel_train).unionAll(water_train)
```

```
#create test set
test_df = bug_test.unionAll(dark_test).unionAll(dragon_test).unionAll(electric_test).unionAll(fairy_test).unionAll(fighting_test).unionAll(fire_test).unionAll(flying_test).unionAll(ghost_test).unionAll(grass_test).unionAll(ground_test).unionAll(ice_test).unionAll(normal_test).unionAll(poison_test).unionAll(psychic_test).unionAll(rock_test).unionAll(steel_test).unionAll(water_test)
```

```
from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline
from sparkdl import DeepImageFeaturizer 

featurizer = DeepImageFeaturizer(inputCol="image", outputCol="features", modelName="InceptionV3")
lr = LogisticRegression(maxIter=15, regParam=0.05, elasticNetParam=0.3, labelCol="label")
p = Pipeline(stages=[featurizer, lr])

p_model = p.fit(train_df)
```

```
tested_df = p_model.transform(test_df)
```

```
tested_df.show()
```

```
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

evaluator = MulticlassClassificationEvaluator(metricName="accuracy")
print("Test set accuracy = " + str(evaluator.evaluate(tested_df.select("prediction", "label"))))
```
